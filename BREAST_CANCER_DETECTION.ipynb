```python
#importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# data collection and processing

# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python


print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# printing the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

1 --> Benign

```python
#Separating the features and target

X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python


print(X)
```

```python


print(Y)
```

```python
#Splitting the data into training data & Testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
# random_state=2 simply makes sure your train/test split is consistent every time you run your code.

```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# === Step 1: Import the StandardScaler from sklearn ===
from sklearn.preprocessing import StandardScaler

# === Step 2: Create a StandardScaler object ===
scaler = StandardScaler()

# === Step 3: Fit the scaler to the training data and transform it ===
# This calculates the mean and standard deviation of X_train,
# then uses those values to scale the training data.
X_train_std = scaler.fit_transform(X_train)

# === Step 4: Transform the test data using the same scaler ===
# This uses the mean and std from X_train to transform X_test.
# We do NOT fit again on X_test to avoid data leakage.
X_test_std = scaler.transform(X_test)
```

```python
#BUILDING A NEURAL NETWORK
# importing tensorflow and Keras
# Set the random seed for TensorFlow to ensure reproducibility
# This makes sure that weight initialization, data shuffling, and other random operations
# behave the same way every time the code is run, which helps in debugging and consistent results.
tf.random.set_seed(3)
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python


# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
#Visualizing accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
#Accuracy of the model on test data


loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
